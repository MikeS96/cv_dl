```
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```
img = cv2.imread('elena_noisy.png')
plt.imshow(img)
plt.show()
```

```
box = np.array(([1, 1, 1], [1, 1, 1], [1, 1, 1]), dtype = 'float') * 1/9
img_wn = cv2.filter2D(img, -1, box)
plt.imshow(img_wn)
plt.show()
```

```
img_gauss = cv2.GaussianBlur(img, (3,3), 0)
plt.imshow(img_gauss)
plt.show()
```

```
plt.subplot(131), plt.imshow(img), plt.title("Noisy lena")
plt.subplot(132), plt.imshow(img_wn), plt.title("Lena box smoothing")
plt.subplot(133), plt.imshow(img_gauss), plt.title("Lena Gaussian")
plt.show()
```

```
img = cv2.imread('ladder.png', 0)
#img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```
# LaPlaciano usando Numpy
laplacian = np.array(([0, 1, 0], [1, -4, 1], [0, 1, 0]), dtype = 'int')
img_laplace = cv2.filter2D(img, -1, laplacian)
cv2_imshow(img_laplace)
```

```
laplaco = cv2.Laplacian(img, cv2.CV_64F)
cv2_imshow(laplaco)
```

```
# Sobel X usando Numpy
sobelX = np.array(([-1, 0, 1], [-2, 0, 2], [-1, 0, 1]), dtype = 'int')
img_sobel_x = cv2.filter2D(img, -1, sobelX)
cv2_imshow(img_sobel_x)
```

```
# Sobel Y usando Numpy
sobelY = np.array(([-1, -2, -1], [0, 0, 0], [1, 2, 1]), dtype = 'int')
img_sobel_y = cv2.filter2D(img, -1, sobelY)
cv2_imshow(img_sobel_y)
```

```
# Multiple LaPlacians usando Numpy
# Reading an image
img_house = cv2.imread('casa.jpg', 0)
l1 = np.array(([0, 1, 0], [1, -4, 1], [0, 1, 0]), dtype = 'int')
l2 = np.array(([1, 1, 1], [1, -8, 1], [1, 1, 1]), dtype = 'int')
l3 = l1 * -1
l4 = l2 * -1
# Computing laplacians
laplace_l1 = cv2.filter2D(img_house, -1, l1)
laplace_l2 = cv2.filter2D(img_house, -1, l2)
laplace_l3 = cv2.filter2D(img_house, -1, l3)
laplace_l4 = cv2.filter2D(img_house, -1, l4)
# showing images
cv2_imshow(laplace_l1)
cv2_imshow(laplace_l2)
cv2_imshow(laplace_l3)
cv2_imshow(laplace_l4)
```
